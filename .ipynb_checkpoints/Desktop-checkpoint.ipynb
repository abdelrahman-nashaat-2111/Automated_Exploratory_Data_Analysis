```python
import random

def create_board():
  """Creates a 3x3 game board."""
  board = []
  for i in range(3):
    row = []
    for j in range(3):
      row.append(' ')
    board.append(row)
  return board

def print_board(board):
  """Prints the game board to the console."""
  for row in board:
    for square in row:
      print(square, end=' ')
    print()

def handle_player_move(board, player):
  """Handles a player's move on the game board."""
  while True:
    print('Player {}'.format(player))
    row = int(input('Enter row: '))
    column = int(input('Enter column: '))
    if 0 <= row < 3 and 0 <= column < 3 and board[row][column] == ' ':
      board[row][column] = player
      break
    else:
      print('Invalid move. Try again.')

def check_for_win(board, player):
  """Checks if the player has won the game."""
  for row in board:
    if all(square == player for square in row):
      return True
  for column in range(3):
    if all(board[row][column] == player for row in range(3)):
      return True
  if board[0][0] == player and board[1][1] == player and board[2][2] == player:
    return True
  if board[0][2] == player and board[1][1] == player and board[2][0] == player:
    return True
  return False

def check_for_tie(board):
  """Checks if the game is a tie."""
  for row in board:
    for square in row:
      if square == ' ':
        return False
  return True

def main():
  """The main game loop."""
  board = create_board()
  player = 'X'
  while True:
    print_board(board)
    handle_player_move(board, player)
    if check_for_win(board, player):
      print('Player {} wins!'.format(player))
      break
    player = 'O' if player == 'X' else 'X'
    if check_for_tie(board):
      print('The game is a tie!')
      break

if __name__ == '__main__':
  main()
```
